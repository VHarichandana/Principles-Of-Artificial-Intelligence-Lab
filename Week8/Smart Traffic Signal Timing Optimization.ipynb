```python
import numpy as np
```

```python
# States: Traffic density levels
states = ["Low", "Medium", "High"]

# Actions: Signal timing decisions
actions = ["Short_Green", "Medium_Green", "Long_Green"]

n_states = len(states)
n_actions = len(actions)
```

```python
# P[action][state][next_state]
P = np.array([
    # Short Green
    [[0.7, 0.3, 0.0],
     [0.2, 0.6, 0.2],
     [0.1, 0.3, 0.6]],

    # Medium Green
    [[0.6, 0.4, 0.0],
     [0.3, 0.5, 0.2],
     [0.2, 0.4, 0.4]],

    # Long Green
    [[0.5, 0.5, 0.0],
     [0.4, 0.4, 0.2],
     [0.3, 0.5, 0.2]]
])
```

```python
# Reward: higher is better (less waiting time)
R = np.array([
    [10, 5, -10],   # Short Green
    [8, 7, -5],     # Medium Green
    [6, 9, 5]       # Long Green
])
```

```python
V = np.zeros(n_states)

gamma = 0.9        # Discount factor
theta = 0.001      # Convergence threshold
```

```python
while True:
    delta = 0
    new_V = np.copy(V)

    for s in range(n_states):
        action_values = []

        for a in range(n_actions):
            value = 0
            for s_next in range(n_states):
                value += P[a][s][s_next] * (R[a][s] + gamma * V[s_next])
            action_values.append(value)

        new_V[s] = max(action_values)
        delta = max(delta, abs(new_V[s] - V[s]))

    V = new_V

    if delta < theta:
        break
```

```python
policy = []

for s in range(n_states):
    action_values = []

    for a in range(n_actions):
        value = 0
        for s_next in range(n_states):
            value += P[a][s][s_next] * (R[a][s] + gamma * V[s_next])
        action_values.append(value)

    best_action = np.argmax(action_values)
    policy.append(actions[best_action])
```

```python
print("Optimal Value Function:")
for i in range(n_states):
    print(f"{states[i]}: {V[i]:.2f}")

print("\nOptimal Policy (Best Signal Timing):")
for i in range(n_states):
    print(f"{states[i]} Traffic → {policy[i]}")
```
