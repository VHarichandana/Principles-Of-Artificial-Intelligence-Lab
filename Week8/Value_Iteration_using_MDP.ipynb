```python
import numpy as np
class GridWorldMDP:
    def __init__(self, size, goal, trap):
        self.size = size
        self.goal = goal
        self.trap = trap
        self.state_space = [(i, j) for i in range(size) for j in range(size)]
        self.action_space = ['UP', 'DOWN', 'LEFT', 'RIGHT']
    def get_next_state(self, state, action):
        i, j = state
        if action == 'UP': i -= 1
        elif action == 'DOWN': i += 1
        elif action == 'LEFT': j -= 1
        elif action == 'RIGHT': j += 1
        i = max(0, min(i, self.size - 1))
        j = max(0, min(j, self.size - 1))
        return (i, j)
    def get_reward(self, state):
        if state == self.goal:
            return 0
        elif state == self.trap:
            return -10
        return -1
def value_iteration(mdp, gamma=0.9, epsilon=0.01):
    V = {s: 0 for s in mdp.state_space}
    while True:
        delta = 0
        new_V = V.copy()
        for s in mdp.state_space:
            if s == mdp.goal or s == mdp.trap:
                continue
            values = []
            for a in mdp.action_space:
                next_s = mdp.get_next_state(s, a)
                reward = mdp.get_reward(next_s)
                values.append(reward + gamma * V[next_s])
            new_V[s] = max(values)
            delta = max(delta, abs(V[s] - new_V[s]))
        V = new_V
        if delta < epsilon:
            break
    return V
```

```python
# Example
mdp = GridWorldMDP(3, (2, 2), (1, 1))
result = value_iteration(mdp)
print("Value Iteration Output:")
for s in sorted(result):
    print(s, ":", round(result[s], 3))
```
