```python
# Leaf node utilities (system security score)
# Higher = more secure (good for Defender, bad for Attacker)

game_tree = {
    "A": ["B", "C"],

    "B": ["D", "E"],
    "C": ["F", "G"],

    "D": [3, 5],
    "E": [6, 9],
    "F": [1, 2],
    "G": [0, -1]
}
```

```python
def minimax(node, depth, alpha, beta, maximizing):

    # If leaf node (integer value)
    if isinstance(node, int):
        return node

    if maximizing:  # Defender (maximize security)
        max_eval = float('-inf')
        for child in game_tree[node]:
            eval = minimax(child, depth+1, alpha, beta, False)
            max_eval = max(max_eval, eval)
            alpha = max(alpha, eval)

            # Alpha-Beta Pruning
            if beta <= alpha:
                break

        return max_eval

    else:  # Attacker (minimize security)
        min_eval = float('inf')
        for child in game_tree[node]:
            eval = minimax(child, depth+1, alpha, beta, True)
            min_eval = min(min_eval, eval)
            beta = min(beta, eval)

            # Alpha-Beta Pruning
            if beta <= alpha:
                break

        return min_eval
```

```python
result = minimax("A", 0, float('-inf'), float('inf'), True)
print("Optimal Security Score:", result)
```

```python
def best_move():
    best_val = float('-inf')
    best_action = None

    for child in game_tree["A"]:
        move_val = minimax(child, 1, float('-inf'), float('inf'), False)

        if move_val > best_val:
            best_val = move_val
            best_action = child

    return best_action, best_val

move, value = best_move()
print("Best Defense Strategy:", move)
print("Expected Security Score:", value)
```
