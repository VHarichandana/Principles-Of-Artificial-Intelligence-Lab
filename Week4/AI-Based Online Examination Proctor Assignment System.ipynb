```python
# Each exam needs one proctor
exams = ["Exam1", "Exam2", "Exam3"]
```

```python
proctors = ["Alice", "Bob", "Charlie"]
```

```python
# Availability: which proctor is available for which exam
availability = {
    "Exam1": ["Alice", "Bob"],
    "Exam2": ["Bob", "Charlie"],
    "Exam3": ["Alice", "Charlie"]
}

# Skill match: proctors qualified for each exam
skills = {
    "Exam1": ["Alice", "Bob"],
    "Exam2": ["Bob", "Charlie"],
    "Exam3": ["Alice", "Charlie"]
}

# Time slots (to avoid conflicts)
time_slots = {
    "Exam1": "10AM",
    "Exam2": "10AM",
    "Exam3": "2PM"
}
```

```python
def is_valid(assignment, exam, proctor):
    # Constraint 1: availability
    if proctor not in availability[exam]:
        return False

    # Constraint 2: skill match
    if proctor not in skills[exam]:
        return False

    # Constraint 3: no same proctor at same time
    for assigned_exam in assignment:
        if assignment[assigned_exam] == proctor:
            if time_slots[assigned_exam] == time_slots[exam]:
                return False

    return True
```

```python
def backtrack(assignment):
    # If all exams assigned → solution found
    if len(assignment) == len(exams):
        return assignment

    # Select unassigned exam
    for exam in exams:
        if exam not in assignment:
            for proctor in proctors:
                if is_valid(assignment, exam, proctor):
                    assignment[exam] = proctor

                    result = backtrack(assignment)
                    if result:
                        return result

                    # Backtrack
                    del assignment[exam]

            return None
```

```python
solution = backtrack({})
```

```python
if solution:
    print("Proctor Assignment:")
    for exam in solution:
        print(f"{exam} → {solution[exam]}")
else:
    print("No valid assignment found")
```
