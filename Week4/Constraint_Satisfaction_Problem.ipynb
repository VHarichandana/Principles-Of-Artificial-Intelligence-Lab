```python
variables = ['A', 'B', 'C']
domains = {
    'A': [1, 2, 3],
    'B': [1, 2, 3],
    'C': [1, 2, 3]
}
constraints = {
    'A': [lambda var, val, ass: 'B' not in ass or ass['B'] != val],
    'B': [lambda var, val, ass: 'A' not in ass or ass['A'] != val],
    'C': [
        lambda var, val, ass: 'A' not in ass or ass['A'] != val,
        lambda var, val, ass: 'B' not in ass or ass['B'] != val
    ]
}
def is_consistent(variable, value, assignment):
    for constraint in constraints.get(variable, []):
        if not constraint(variable, value, assignment):
            return False
    return True
def backtrack(assignment):
    if len(assignment) == len(variables):
        return assignment
    for var in variables:
        if var not in assignment:
            break
    for value in domains[var]:
        if is_consistent(var, value, assignment):
            assignment[var] = value
            result = backtrack(assignment)
            if result is not None:
                return result
            assignment.pop(var)
    return None
solution = backtrack({})
if solution:
    print("Solution found:")
    for v in solution:
        print(v, ":", solution[v])
else:
    print("No solution found")
```
